```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./full_data.csv')
```

```python
df.columns
```

```python
goal_data = df[['Round', 'Date', 'H_Score', 'A_Score']]


goals_per_round = goal_data.groupby('Round')['H_Score', 'A_Score'].sum().reset_index()
goals_per_date = goal_data.groupby('Date')['H_Score', 'A_Score'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(goals_per_round['Round'], goals_per_round['H_Score'], label='Home Team Goals')
plt.plot(goals_per_round['Round'], goals_per_round['A_Score'], label='Away Team Goals')
plt.xlabel('Round')
plt.ylabel('Number of Goals')
plt.title('Number of Goals per Round')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.plot(goals_per_date['Date'], goals_per_date['H_Score'], label='Home Team Goals')
plt.plot(goals_per_date['Date'], goals_per_date['A_Score'], label='Away Team Goals')
plt.xlabel('Date')
plt.ylabel('Number of Goals')
plt.title('Number of Goals per Date')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### Do teams with more goal attempts or shots on goal have a higher chance of winning? Is there a difference between home and away teams in this regard?

```python
relevant_columns = ['H_Goal_Attempts', 'A_Goal_Attempts', 'H_Shots_on_Goal', 'A_Shots_on_Goal', 'WIN', 'H_Score', 'A_Score']
data = df[relevant_columns]

data['Total_Goal_Attempts'] = data['H_Goal_Attempts'] + data['A_Goal_Attempts']
data['Total_Shots_on_Goal'] = data['H_Shots_on_Goal'] + data['A_Shots_on_Goal']
data['Avg_Goal_Attempts'] = data['Total_Goal_Attempts'] / 2
data['Avg_Shots_on_Goal'] = data['Total_Shots_on_Goal'] / 2


outcome_metrics = data.groupby('WIN')['Avg_Goal_Attempts', 'Avg_Shots_on_Goal'].mean()


home_metrics = data[data['WIN'] == 'H'].groupby('WIN')['Avg_Goal_Attempts', 'Avg_Shots_on_Goal'].mean()
away_metrics = data[data['WIN'] == 'A'].groupby('WIN')['Avg_Goal_Attempts', 'Avg_Shots_on_Goal'].mean()


print("Overall Metrics:")
print(outcome_metrics)
print("\nHome Team Metrics:")
print(home_metrics)
print("\nAway Team Metrics:")
print(away_metrics)
```

```python

columns = ['H_Fouls', 'A_Fouls', 'H_Yellow_Cards', 'A_Yellow_Cards', 'Home', 'Away']
data = df[columns]


data['Total_Fouls'] = data['H_Fouls'] + data['A_Fouls']
data['Total_Yellow_Cards'] = data['H_Yellow_Cards'] + data['A_Yellow_Cards']
data['Avg_Fouls_per_Match'] = data['Total_Fouls'] / 2
data['Avg_Yellow_Cards_per_Match'] = data['Total_Yellow_Cards'] / 2


team_stats = data.groupby(['Home', 'Away'])[['Avg_Fouls_per_Match', 'Avg_Yellow_Cards_per_Match']].mean()


teams_with_high_fouls = team_stats.sort_values('Avg_Fouls_per_Match', ascending=False)
teams_with_high_yellow_cards = team_stats.sort_values('Avg_Yellow_Cards_per_Match', ascending=False)



print("Teams with higher average fouls per match:")
print(teams_with_high_fouls.head())

print("\nTeams with higher average yellow cards per match:")
print(teams_with_high_yellow_cards.head())



plt.figure(figsize=(10, 6))
plt.bar(team_stats.index, team_stats['Avg_Fouls_per_Match'])
plt
```
